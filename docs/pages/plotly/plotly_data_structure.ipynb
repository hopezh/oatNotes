# show data strucure of a plotly figure

# example figure

````python
import plotly.express as px

fig = px.line(
    x=["a","b","c"],
    y=[1,3,2],
    title="sample figure"
)

fig.show()
````

# 3 ways to view a fig’s data strucure

As a verbose `layout.template` key (shown as ellipses `...` for
brevity).

``` python
print(fig)      # opt 1
# fig.to_dict()   # opt 2
# fig.to_json()   # opt 3
```
